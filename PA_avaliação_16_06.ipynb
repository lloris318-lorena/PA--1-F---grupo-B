```python
# QUESTÃO 01 – Contagem Crescente (2 pontos)
# Elabore um programa que exiba na tela os números de 1 até 20,
# um em cada linha.

# Exemplo de saída:

# 1
# 2
# 3
# ...
# 20

for i in range (1,21):
  print (i)
```

```python
# QUESTÃO 02 – Soma dos Números (2 pontos)

# Solicite ao usuário um número inteiro positivo.

# Utilizando um laço de repetição, calcule e exiba a soma de
# todos os números de 1 até o número informado.

# Exemplo:

# Digite um número: 5

# Resultado: 15

# Pois:

# 1 + 2 + 3 + 4 + 5 = 15

numero = int(input("Digite um número inteiro e positivo."))

soma = 0

for i in range(1,numero + 1):
  soma += i
print(f"A soma dos numeros até {numero} é: {soma}")
```

```python
# QUESTÃO 03 – Tabuada (2 pontos)

# Solicite ao usuário um número inteiro e exiba sua tabuada de 1 até 10.

# Exemplo:

# Digite um número: 7

# 7 x 1 = 7
# 7 x 2 = 14
# 7 x 3 = 21
# ...
# 7 x 10 = 70

numero = int (input("digite um numerode 1 a 10."))

for i in range (1,11):
  print (f"{numero} x {i} = {numero * i }")
```

```python
# QUESTÃO 04 – Fatorial (4 pontos)

#Solicite ao usuário um número inteiro positivo e calcule seu fatorial utilizando um laço de repetição.

# Exemplo:

# Digite um número: 5

# Fatorial = 120

# Pois:

# 5! = 5 × 4 × 3 × 2 × 1

# Observação:
# O programa deve validar se o número informado é positivo.
# Solicita o número e garante que ele seja um inteiro positivo


numero = int(input("Digite um número inteiro positivo: "))

while numero < 0:
    print("Valor inválido! Digite um número inteiro positivo.")
    numero = int(input("Digite um número inteiro positivo: "))

fatorial = 1

for i in range(1, numero + 1):
    fatorial *= i

print(f"Fatorial de {numero} =", fatorial)
```

```python
# QUESTÃO 05 – Caixa Eletrônico (5 pontos)


# Um caixa eletrônico possui notas de:

# R$ 100
# R$ 50
# R$ 20
# R$ 10
# R$ 5
# R$ 2
# R$ 1

# Solicite ao usuário o valor do saque e informe a menor quantidade possível de notas para realizar o pagamento.

# Exemplo:

# Digite o valor do saque: 287

# Saída esperada:

# 2 nota(s) de R$100
# 1 nota(s) de R$50
# 1 nota(s) de R$20
# 1 nota(s) de R$10
# 1 nota(s) de R$5
# 1 nota(s) de R$2

# Requisitos:
# - Utilizar laços de repetição.
# - Não utilizar listas.
# - Exibir apenas as notas utilizadas.

valor = int (input("digite um valor:"))

qtd100 = 0
while valor >= 100:
    valor -= 100
    qtd100 +=1

qtd50 = 0
while valor >= 50:
    valor -= 50
    qtd50 +=1

qtd10 = 0
while valor >= 10:
    valor -= 10
    qtd10 +=1

qtd5 = 0
while valor >= 5:
    valor -= 5
    qtd5 +=1

qtd2 = 0
while valor >= 2:
    valor -= 2
    qtd2 +=1

print(f"notas de 100: {qtd100}")
print(f"notas de 50: {qtd50}")
print(f"notas de 10: {qtd10}")
print(f"notas de 2: {qtd2}")
print(f"notas de 5:{qtd5}")
```
