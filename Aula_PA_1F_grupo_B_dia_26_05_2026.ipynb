```python
# projeto caixa eletrônico

# Você foi para desenvolver um sistema simples de caixa eletrônico.

# O sistema deverá funcionar através de um menu interativo utilizando "while"

# O programa deve iniciar com um saldo inicial  de R$1000,00 e permitir que  o usuário escolha diferentes operações até decidir sair do sistema.

# Ver saldo
# Depositar
# Sacar Dinheiro
# Sair

# Regras:

# Quando depositar somar ao saldo anterior;
# Quando sacar subtrair do saldo anterior;
# Mostrar o saldo para o usuário
```

```python
print("======MENU======")
print ("")
print("Opção 1 - Ver saldo ")
print("Opção 2 - Depositar ")
print("Opção 3 - Sacar Dinheiro")
print("Opção 4 - Sair ")
print("================")

saldo = 1000.00
opcao = ""

while opcao != 4:
  try:

    opcao = int(input("Escolha a opção desejada: "))

    if opcao == 1:
        print(f"O seu Saldo atual é de: R$ {saldo}")

    elif opcao == 2:
        deposito = float(input("Digite o valor que deseja depositar? "))
        saldo =  saldo + deposito
        print ("")

    elif opcao == 3:
      deposito = float (input("Digite o valor que deseja sacar:"))
      saldo = saldo - deposito
      print ("")

    elif opcao < 0 or opcao > 3:
      print ("Escolha uma opção")
  except:
    print ("Escolha uma opção de 1 a 3 ou 4 para sair!")
```
