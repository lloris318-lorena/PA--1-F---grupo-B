```python
#1.	Par ou ímpar Peça um número ao usuário e informe se ele é par ou ímpar.

#entrada

numero = float (input("digite um numero"))

#processamento e saida

if numero % 2 == 0:
  print(f"O numero{numero} é Par.")

else:
  print (f"o numero {numero} é impar.")
```

```python
#2.	Maior de dois números Solicite dois números e mostre qual é o maior (ou se são iguais).

#entrada

numero1 = float (input("digite um numero"))
numero2 = float (input("digite um numero"))

if numero1 > numero2:
  print (f"{numero1} é maior que o {numero2}.")

else:
  print (f"{numero2} é maior que o {numero1}.")
```

```python
#3.	Aprovação de aluno Peça a nota de um aluno e informe:  o	Aprovado (nota ≥ 7) o	Recuperação (5 ≤ nota < 7)  o	Reprovado (nota < 5)

#entrada
nota = float (input("digite sua nota."))

#processamento e saida

if nota >= 7:
  print (f"Aprovado")

elif nota >= 5  and nota <7:
  print (f"Recuperação")

else:
  print (f"Reprovado")
```

```python
#4.	Classificação de idade Receba a idade e classifique: o	Criança (0–12) o	Adolescente (13–17) o	Adulto (18–59) o	Idoso (60+)

#entrada

idade = int (input("digite sua idade."))

#processamento e saida

if idade >= 0 and idade <= 12:
  print (f"Você é criança.")

elif idade >= 13 and idade <=17:
  print (f"Você é adolescente.")

elif idade >=18 and idade <=59:
  print (f"Você é Adulto.")

else:
  print  (f"Você é idoso.")

```

```python
#6.	Positivo, negativo ou zero Peça um número e diga se ele é positivo, negativo ou zero.

#Entrada

numero = float (input("digite um numero"))

#processamento e Saida

if numero < 0:
  print(f"negativo")

elif numero > 0:
  print(f"esse numero e positivo")

else:
  print(f" Esse numero é zero.")
```

```python
#7.Calculadora simples Peça dois números e uma operação (+, -, *, /). Use if/elif para calcular o resultado.

#entrada

num1 = float (input("digite um numero."))
num2 = float (input("digite outro numero."))
operacao = str (input("digite uma operação (exemplo: +,-, *, /)."))
soma = num1 + num2
subtracao = num1 - num2
multiplicacao = num1 * num2
divisao = num1 / num2

#Processamento e saida

if operacao == "+":
  print(f"{soma}")

elif operacao == "-":
  print(f"{subtracao}")

elif operacao == "*":
  print(f"{multiplicacao}")

else:
  print(f"{divisao}")
```
