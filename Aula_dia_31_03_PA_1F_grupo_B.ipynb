```python
#Entrada

numero1 = int (input ("digite um numero"))
numero2 = int (input ("digite um numero"))

#processamento e saida

if numero1 > numero2:
  print (f"O número {numero1} é maior que o {numero2}.")

elif numero1 == numero2:
  print (f"O número {numero1} é igual {numero2}")

else:
  print(f"O número {numero1} é menor que o {numero2}.")
```

```python
# Escrever um programa que o usuario digita o nome e a idade:

# Indentificar se o usuario é criança, adolesacente ou adulto.

# < 12 Criança

# < 18 Adolescente

# < 18 Adulto

# saida ( Você é ...)

# Entrada

nome =  str (input ("Digite seu nome:"))
idade = int (input ("Digite sua idade:"))

#Processamento/Saida

if idade <= 12:
  print (f"{nome},  Você tem {idade}, Você é uma Criança.")

elif idade <= 18:
  print (f" {nome}, Você tem {idade}, Você é um Adolescente.")

else:
  print (f"{nome}, Você tem {idade}, Você é um adulto.")
```
