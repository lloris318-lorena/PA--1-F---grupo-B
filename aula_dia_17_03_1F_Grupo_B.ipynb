```python
#Operadores aritiméticos

#Pedir ao usuario para digitar 02 numeros

#Entrada

a = int (input("digite um numero"))
b = int (input("digite um numero"))

#processamento

#Soma

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
div_int = a // b
resto = a % b
potencia = a ** b

#Saida

print("")
print("----------OPERADORES ARITIMETICOS--------")
print("")
print("-------------------SOMA-------------")
print("")
print(f"O resultado da soma de {a} mais {b} é igual a :{soma}")
print("---------------------------")


print("-------------------SUBTRAÇÃO-------------")
print("")
print(f"O resultado da subtração de {a} menos {b} é igual a :{subtracao}")
print("------------------------------")

print("-------------------MULTIPLICAÇÃO-------------")
print("")
print(f"O resultado da multiplição de {a} vezes {b} é igual a :{multiplicacao}")
print("------------------------------")

print("-------------------DIVISÃO-------------")
print("")
print(f"O resultado da divisão de {a} dividido {b} é igual a :{divisao}")
print("------------------------------")

print("-------------------DIVISÃO INTEIRA-------------")
print("")
print(f"O resultado da divisão inteira de {a} dividido {b} é igual a :{div_int}")
print("------------------------------")

print("-------------------RESTO-------------")
print("")
print(f"O resto da divisão de {a} dividido {b} é igual a :{resto}")
print("------------------------------")

print("-------------------POTENCIA-------------")
print("")
print(f"O resultado da potencia de {a} elevado {b} é igual a :{potencia}")
print("------------------------------")



```
